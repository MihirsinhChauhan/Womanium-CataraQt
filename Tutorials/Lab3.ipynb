# Visualization on a Bloch Sphere

In the previous lab we have seen Pauli operators. Now let's see them on a bloch sphere.

- Hadamard gate creates a superposition which can be seen by the state represented by a red arrow in the below diagram.

```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit, execute, Aer

mycircuit = QuantumCircuit(qreg)

qreg =  QuantumRegister(1)
creg = ClassicalRegister(1)

mycircuit = QuantumCircuit(qreg,creg)


mycircuit.h(qreg[0])


job = execute(mycircuit,Aer.get_backend('statevector_simulator'),optimization_level=0)
current_quantum_state=job.result().get_statevector(mycircuit)

%run quantum.py
angles = state_to_angles(current_quantum_state)

%run drawing.py
draw_bloch(angles[0], angles[1])
```

- Below is an implementation of Z gate.

```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit, execute, Aer

mycircuit2 = QuantumCircuit(qreg)

qreg =  QuantumRegister(1)
creg = ClassicalRegister(1)

mycircuit2 = QuantumCircuit(qreg,creg)


mycircuit2.z(qreg[0])


job = execute(mycircuit2,Aer.get_backend('statevector_simulator'),optimization_level=0)
current_quantum_state=job.result().get_statevector(mycircuit2)

%run quantum.py
angles = state_to_angles(current_quantum_state)

%run drawing.py
draw_bloch(angles[0], angles[1])
```

We can see there is no change of state happen here in the above diagram even after applying Z operator. "minus" is the eigenvalue of the Z operator.

- Let's create a superposition and then apply a  Z  gate. Note that superposition is created by applying hadamard or H gate.

```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit, execute, Aer

mycircuit3 = QuantumCircuit(qreg)

qreg =  QuantumRegister(1)
creg = ClassicalRegister(1)

mycircuit3 = QuantumCircuit(qreg,creg)


mycircuit3.h(qreg[0])
mycircuit3.x(qreg[0])

mycircuit3.z(qreg[0])



job = execute(mycircuit3,Aer.get_backend('statevector_simulator'),optimization_level=0)
current_quantum_state=job.result().get_statevector(mycircuit3)

%run quantum.py
angles = state_to_angles(current_quantum_state)

%run drawing.py
draw_bloch(angles[0], angles[1])
```

Initial state: |0>

after applying X gate: |1>

after applying H gate: |0> + |1>/ sqrt 2

after applying Z gate: |0> - |1>/ sqrt 2
